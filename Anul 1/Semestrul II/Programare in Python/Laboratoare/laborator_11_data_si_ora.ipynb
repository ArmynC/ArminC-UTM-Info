## timedelta

datele si orele au propria aritmetica. timedelta reprezinta "trecerea timpului". Timedelta este in acelasi timp o functie care permite adaugarea (sau scaderea) intervalelor din date sau un obiect care reprezinta aceasta diferenta. 

```python
from datetime import timedelta

print(ndt)

dd = ndt + timedelta(days=3)
print(dd)

print(ndt)
```

#Problema 1

Cate luni au trecut de la craciun?

Cate luni au trecut de la cel mai recent craciun?

```python
current_date = datetime.now()

target_date = datetime(current_date.year, 12, 25)

if current_date > target_date:
  months = (current_date.year - target_date.year) * 12 + (current_date.month - target_date.month)
else:   
  target_date = datetime(current_date.year-1, 12, 25)
  months = (current_date.year - target_date.year) * 12 + (current_date.month - target_date.month)


print(months)  
```

#Problema 2

Scrieti un program care afiseaza toate duminicile dintr-un anumit an

```python
anul_curent = 2024

prima_zi = date(anul_curent,1,1)
prima_duminica = prima_zi + timedelta(days = 6 - prima_zi.weekday())

duminica_curenta = prima_duminica

while duminica_curenta.year == anul_curent:
   print(duminica_curenta)
   duminica_curenta = duminica_curenta + timedelta(days = 7)
```

# Problema 3

Scrieti un program care sa testeze daca o anumita data trimisa ca parametru de ti sir este cea de-a treia marti din luna respectiva.

```python
def a_treia_marti(sir):
  dat = datetime.strptime(sir, '%Y-%m-%d')
  rez = 14 < dat.day < 22 and dat.weekday() == 1
  return rez

test = a_treia_marti('2023-05-16') # true
print(test)

test2 = a_treia_marti('2022-05-16') # false
print(test2)

test3 = a_treia_marti('2024-04-16') # false
print(test3)

```

#Problema 4

Scrieti un program care sa numere zilele de luni care pica in prima zi a unei luni intre anii 2015 - 2023

```python
lunea = 0

for year in range(2015, 2024):
  for month in range(1, 13):
    if datetime(year,month, 1).weekday() == 0:
      lunea += 1

print(lunea)
```
