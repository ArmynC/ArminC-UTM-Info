Scrieti o functie care testeaza daca un numar este prim. Functia primeste ca argument numarul respectiv si returneaza True daca numerul este prim si False daca numarul nu este prim.

- daca n < 1 -> retunr false
- testam toate numerele de la 2 la n-1 si vedem daca n % i == 0
- daca n % i == 0 -> numarul n se imparte exact la i deci nu este prim



````python
#Horia Banica

def is_prime_h(x):
  if x < 2:
    return False
    
  i = 2
  while i * i <= x:
    if x % i == 0:
      return False
    i += 1
  return True

````

````python
print(is_prime_h(20))
````

Scrieti o functie care returneaza toate numerele prime mai mici decat un numar intgreg trimis ca parametru

```
print(nr_prime(20)) 
[2, 3, 5, 7, 11, 13, 17, 19]
```


````python
def nr_prime(n):
  prime = []
  for num in range(2,n+1):
    for i in range(2, num-1):
      if num % i == 0:
        break
    else:
      prime.append(num)
  return prime


print(nr_prime(20))
````

Putem apela functia deja construita de testare a caracterului prim al unui numar (is_prime_h)

````python
def nr_prime_2(n):
  prime = []
  for num in range(2,n+1):
    if is_prime_h(num):
      prime.append(num)

  return prime

print(nr_prime_2(20))
````

Extindeti functia nr_prime_2 astfel incat sa primeasca doua argumente si sa testeze numerele prime din intervalul definit de cele doua numere

````python
def nr_prime_2_in(m, n):
  prime = []
  for num in range(m,n+1):
    if is_prime_h(num):
      prime.append(num)

  return prime

print(nr_prime_2_in(20, 100))
````

scrieti o functie care inlocuieste toate vocalele dintr-un sir cu alte vocale

```
print(inlcvoc("Hello World"))
Hilli wirld

```

````python
import random

def inlcvoc(sir):
  vocale = "aeiou"
  result = ""
  lv = random.choice(vocale)
  for char in sir:
    if char in vocale:
      result += lv
    else:
      result += char

  return result


````

````python
print(inlcvoc("Hello world"))
````

Codul de mai sus alege o singura vocala pentru a inlocui toate vocalele existente. Modificati programul astfel incat la fiecare vocala sa se aleaga alta vocala alternativa

````python
def inlcvoc2(sir):
  vocale = "aeiou"
  result = ""
  for char in sir:
    if char in vocale:
      lv = random.choice(vocale)
      result += lv
    else:
      result += char

  return result
````

````python
print(inlcvoc2("Hello world"))
````

Folositi functia "vocala" definita mai sus

````python
def pasareasca(word):
  result = ""
  i=0
  while i < len(word):
    if (vocala(word[i])):
      i+=1
    else:
      start=i
      while i < len(word) and not vocala(word[i]):
        i+=1
        grup = word[start:i+1]
        result = result + grup + grup
  
  return result
````

````python
# Horia Banica

def aritmetica(a, b, op = "add"):
  if op == "add":
    return a + b
  elif op == "div":
    return a / b
  elif op == 'sub':
    return a - b
  elif op == 'mul':
    return a * b
  else:
    print("Invalid op. Choose add, sub, mul or div")
````

Numar variabil de parametrii de tip keyword

````python
def calculate_total_cost(price, **kwargs):
    tax_rate = kwargs.get('tax_rate', 0.05)
    discount = kwargs.get('discount', 0)
    shipping_cost = kwargs.get('shipping_cost', 5.0)
    light_tax = kwargs.get('light_tax', 1.0)

    total_cost = price + (price * tax_rate) - discount + shipping_cost + light_tax
    return total_cost

# Example usage
price = 100.0

total_cost = calculate_total_cost(price, tax_rate=0.08, discount=10.0, shipping_cost=8.0, light_tax=1.2)

print(f"Total cost: {total_cost}")
````
